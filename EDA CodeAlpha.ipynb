```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Sales'].describe()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Sales'], bins=50, color='steelblue')
plt.title('Distribution of Sales')
plt.xlabel('Sales($)')
plt.ylabel('Count')
plt.show()
```

```python
df['Profit'].describe()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Profit'], bins=50, color='blue')
plt.title('Distribution of Profit')
plt.xlabel('Profit($)')
plt.ylabel('Count')
plt.show()
```

```python
df['Cost'] = df['Sales'] - df['Profit']
```

```python
df[['Product Name','Sales', 'Profit', 'Cost']].head(10)
df[['Product Name','Sales', 'Profit', 'Cost']].tail(10)
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='muted')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(category_profit)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=category_profit.index, y=category_profit.values, palette='muted')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Peofit ($)')
plt.show()
```

```python
category_discount = df.groupby('Category')['Discount'].mean().sort_values()
print(category_discount)
```

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values()
print(region_profit)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=region_profit.index, y=region_profit.values, palette='muted')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['Discount'], y=df['Profit'], alpha=0.3, color='steelblue')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit ($)')
plt.axhline(0, color='red', linewidth=1, linestyle='--')
plt.show()
```

## Key Insights — Superstore EDA

### 1. Sales
- Technology is the top-selling category with $836K
- Most orders have a small value — below $54

### 2. Profitability
- Technology is the most profitable category with $145K
- Furniture has $742K in sales but only $18K in profit

### 3. Discounts
- Furniture has the highest average discount at 17.3% which explains its low profit
- Discounts above 40% turn orders into losses

### 4. Regions
- West is the most profitable region with $108K
- Central is the weakest region with only $39K
